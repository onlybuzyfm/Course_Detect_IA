```python
from PIL import Image
from PIL.ExifTags import TAGS

img = Image.open("47440.jpg")
exif_data = img.getexif()

for tag, val in exif_data.items():
    nombre = TAGS.get(tag, tag)
    print(f"{nombre}: {val}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("47440.jpg", 0)  # escala de grises
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitud = 20 * np.log(np.abs(fshift))

plt.imshow(magnitud, cmap='gray')
plt.title("Frecuencia (FFT)")
plt.axis('off')
plt.show()
```
